<span style="color:orange">llista.reverse()</span>

```python
llista = [23, 45, 56, 7]

print(llista.reverse())

print(llista)
```

<span style="color:orange">llista.sort()</span>

```python
llista = [23, 45, 56, 7]

print(llista.sort())

print(llista)
```

<span style="color:orange">llista.sort(reverse = True)</span>

```python
llista = [23, 45, 56, 7]

print(llista.sort(reverse = True))

print(llista)
```
